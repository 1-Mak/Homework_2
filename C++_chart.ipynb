```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Data_sorts.xlsx')
```

```python
display(data)
```

```python
index = data['Nums']
nums = data['index']
merge_sort = data['Merge_sort']
quick_sort = data['Quick_sort']
bubble_sort = data['Bubble_sort']
plt.figure(figsize=(15, 5))
plt.plot(nums, merge_sort, label='Merge Sort', marker='o')
plt.plot(nums, quick_sort, label='Quick Sort', marker='o')
plt.plot(nums, bubble_sort, label='Bubble Sort', marker='o')
plt.xticks(nums, index)
plt.yscale('log')
plt.xlabel('Number of elements  (2^x)')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Sorting Algorithms')
plt.legend()
plt.grid(True)
```
